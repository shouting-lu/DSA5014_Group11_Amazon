```python
import os
import pandas as pd
import json
import time
from datetime import datetime, timedelta
import random
import string
import numpy as np
```

```python
data_file = open('Clothing_Shoes_and_Jewelry.json')
data = []

start_time = time.time()
i = 0
for line in data_file:
    i = i + 1
    data.append(json.loads(line))
    if i > 10:
        break
end_time = time.time()
print(end_time-start_time)
df = pd.DataFrame(data)
data_file.close()
```

```python
def generate_review_json(df):
    feas = ['asin', 'reviewerID', 'reviewerName', 'overall', 'vote', 'reviewTime', 'style', 'summary', 'reviewText', 'image']
    df = df[feas]
    df.rename(columns = {'reviewerID' : 'customerID', 'reviewerName': 'customerName'}, inplace = True)
    df['vote'].fillna(0, inplace=True)

    # process date
    date = []
    for date_str in list(df['reviewTime']):
        date_obj = datetime.strptime(date_str, "%m %d, %Y")
        date.append(date_obj.strftime("%Y-%m-%d"))

    df['reviewTime'] = date

    return df
```

```python
review_json = generate_review_json(df)
```
