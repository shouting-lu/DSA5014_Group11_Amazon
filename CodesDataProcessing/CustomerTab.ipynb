```python
import os
import pandas as pd
import json
import time
from datetime import datetime, timedelta
import random
import string
import names
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/5104/ReviewInfo_final.csv")
df.head()
```

```python
def generate_random_string(length):
    characters = string.ascii_letters + string.digits  # 字母和数字
    return ''.join(random.choice(characters) for _ in range(length))

def generate_customer_json(df_reviewer):
    num_strings = len(df_reviewer)
    string_length = 14

    random_ids = [str.zfill(string_length) for _ in range(num_strings)]
    random_names = [names.get_full_name() for _ in range(num_strings)]
    column_names = ['customerID', 'customerName']
    df_customer = pd.DataFrame(list(zip(random_ids, random_names)), columns=column_names)

    result = pd.concat([df_reviewer, df_customer[~df_customer['customerID'].isin(df_reviewer['customerID'])]],
                       ignore_index=True)

    return result
```

```python
df_reviewer = df[['customerID', 'customerName']]
num_strings = len(df_reviewer)
string_length = 14
random_ids = [str(_).zfill(string_length) for _ in range(num_strings)]
random_ids
```
