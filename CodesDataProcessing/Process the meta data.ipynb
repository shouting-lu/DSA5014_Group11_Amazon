```python
import json
import pandas as pd
import time
```

```python
import json

# 指定 JSON 文件的路径
json_file_path = 'meta_Clothing_Shoes_and_Jewelry.json'

# 初始化数据记录数
num_records = 0

# 打开 JSON 文件逐行读取数据
with open(json_file_path, 'r', encoding='utf-8') as json_file:
    for line in json_file:
        try:
            # 尝试解析当前行的 JSON 对象
            data = json.loads(line)
            num_records += 1
        except json.JSONDecodeError:
            # 如果行无法解析为 JSON 对象，跳过该行
            continue

print(f"数据记录数：{num_records} 条")
```

```python
#data_file = open('Clothing_Shoes_and_Jewelry.json')
data_file = open('meta_Clothing_Shoes_and_Jewelry.json')
data = []

start_time = time.time()
i = 0
for line in data_file:
    i = i + 1
    data.append(json.loads(line))
    if i > 2685058:
        break
end_time = time.time()
print(end_time-start_time)

data_file.close()
df = pd.DataFrame(data)
```

```python
df
```

## product


```python
filtered_data = []
for item in data:
    filtered_item = {
        "asin": item.get("asin"),
        "title": item.get("title"),
        "brand": item.get("brand"),
        "p_rank": item.get("rank"),        
        "rate": item.get("date"), 
        "price": item.get("price"),    
        "description": item.get("description"),
        "image": item.get("imageURLHighRes")
    }
    filtered_data.append(filtered_item)

product = pd.DataFrame(filtered_data)
```

```python
#product['rate'] = product['rate'].apply(lambda x: x if 'star' in str(x) else None)
import random
product['rate'] = product['rate'].apply(lambda x: 0 if 'star' not in str(x) else x)
product['rate'] = product['rate'].str.replace(r'5 star5 star \(0%\)', '5', regex=True)
product['rate'] = product['rate'].str.replace(r'5 star', '5', regex=True)

product['rate'] = product['rate'].apply(lambda x: random.randint(1, 4) if ('5') not in str(x) else x)
#product['rate'] = product['rate'].fillna('None')
```

```python
filtered_rows = product[product['rate']==2]
filtered_rows
```

```python
product.to_csv("product.csv")
```
